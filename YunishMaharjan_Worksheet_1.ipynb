```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = "cameraman.png"
img = Image.open(image_path).convert("L")
img_array = np.array(img)

# Task 1: Display the grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(img_array, cmap="gray")
plt.title("Original Grayscale Image")
plt.axis("off")
plt.show()
```

```python
# Task 2: Extract and display the middle 150×150 pixel section
height, width = img_array.shape
start_x = (width // 2) - 75
start_y = (height // 2) - 75
middle_section = img_array[start_y:start_y + 150, start_x:start_x + 150]

plt.figure(figsize=(6, 6))
plt.imshow(middle_section, cmap="gray")
plt.title("Middle 150×150 Pixel Section")
plt.axis("off")
plt.show()
```

```python
# Task 3: Apply a threshold (set values <100 to 0, others to 255)
thresholded_img = np.where(img_array < 100, 0, 255)

plt.figure(figsize=(6, 6))
plt.imshow(thresholded_img, cmap="gray")
plt.title("Thresholded Binary Image")
plt.axis("off")
plt.show()
```

```python
# Task 4: Rotate the image 90 degrees clockwise
rotated_img = np.rot90(img_array, k=-1)

plt.figure(figsize=(6, 6))
plt.imshow(rotated_img, cmap="gray")
plt.title("Rotated Image (90° Clockwise)")
plt.axis("off")
plt.show()
```

```python
# Task 5: Convert grayscale image to RGB (replicating the gray values)
rgb_img = np.stack((img_array,) * 3, axis=-1)

plt.figure(figsize=(6, 6))
plt.imshow(rgb_img)
plt.title("Grayscale Converted to RGB")
plt.axis("off")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy.linalg import eig

# Load the image (Ensure "Lenna.png" is uploaded)
image_path = "Lenna.png"
img = Image.open(image_path).convert("L")
img_array = np.array(img)

# Display Original Image
plt.figure(figsize=(6, 6))
plt.imshow(img_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()

# Step 1: Center the dataset (Standardization)
mean_img = np.mean(img_array, axis=0)
centered_img = img_array - mean_img
```

```python
# Step 2: Compute Covariance Matrix
cov_matrix = np.cov(centered_img, rowvar=False)
```

```python
# Step 3: Compute Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = eig(cov_matrix)

# Sort Eigenvalues & Eigenvectors in Descending Order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
# Step 4: Select Different k Values
k_values = [10, 20, 50, 100, 150]
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

# Show Original Image
axes[0, 0].imshow(img_array, cmap="gray")
axes[0, 0].set_title("Original")
axes[0, 0].axis("off")

# Loop Through k-values and Show Compressed Images
for i, k in enumerate(k_values):
    top_k_eigenvectors = eigenvectors[:, :k]
    transformed_img = np.dot(centered_img, top_k_eigenvectors)
    reconstructed_img = np.dot(transformed_img, top_k_eigenvectors.T) + mean_img

    # Ensure values are within [0, 255] and convert to uint8
    reconstructed_img = np.clip(reconstructed_img, 0, 255).astype(np.uint8)

    # Find correct subplot position (row, col)
    row, col = divmod(i + 1, 3)
    axes[row, col].imshow(reconstructed_img, cmap="gray")
    axes[row, col].set_title(f"k = {k}")
    axes[row, col].axis("off")

plt.tight_layout()
plt.show()
```
